```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
linx = np.linspace(-5, 5, 10001)
sine = np.sin(linx)
plt.plot(linx, sinus)
```

```python
x = np.c_[linx]
y = np.c_[sine]
```

```python
from sklearn.neural_network import MLPRegressor

reg = MLPRegressor(hidden_layer_sizes=(5000,), activation='relu')

reg.fit(x,y)
```

```python
y_p = reg.predict(x)
```

```python
plt.plot(x, y)
plt.plot(x, y_p)
```
